## Working of a Neuron in Artificial Neural Networks (ANNs)

A **neuron** in an ANN is very similar to a perceptron with one key difference:  

- In a perceptron, the **activation function** is a **step function**.  
- In a neuron, the activation function can be **any non-linear function** (e.g., Sigmoid, ReLU, Tanh).  

---

### 1. Structure of a Neuron

1. **Inputs (x, x, ..., x)**  
   - Each input represents a feature or signal from the previous layer.

2. **Weights (w, w, ..., w)**  
   - Each input is multiplied by a corresponding weight representing its importance.

3. **Bias (b)**  
   - Added to the weighted sum to shift the activation function.

4. **Weighted Sum (z)**  
\[
z = w_1 x_1 + w_2 x_2 + ... + w_n x_n + b
\]

5. **Activation Function (f)**  
   - Converts the weighted sum into an output:
\[
y = f(z)
\]

---

### 2. Layers in a Neural Network

- **Input Layer:**  
  - Number of neurons = number of input features.
- **Hidden Layers:**  
  - Can have one or more layers, number of neurons varies.
- **Output Layer:**  
  - Number of neurons = number of classes (for classification)  
  - Number of neurons = 1 (for regression)

---

### 3. Complete Network Specification

To fully define a neural network, we need:

1. **Network Topology**  how many layers, and number of neurons per layer  
2. **Input Layer**  size = number of features  
3. **Output Layer**  size = number of target classes or 1 for regression  
4. **Weights**  initialized and adjusted during training  
5. **Biases**  one per neuron, adjusted during training  
6. **Activation Functions**  one per neuron, defines output non-linearity

## Inputs and Outputs of a Neural Network

### 1. Input Layer

- The number of neurons in the **input layer** = number of input features.
- **Inputs must be numeric**.
- Different types of data are converted to numeric form:
  - **Text data:**  
    - One-hot vectors or word embeddings  
    - Example: Vocabulary size |V|  word w  vector of size |V| with 1 at n position  
  - **Image data:**  
    - Images are arrays of pixels  
    - Example: MNIST grayscale image (18  18)  vector of size 324  
    - RGB images  number of neurons = height  width  3  

---

### 2. Output Layer

- Number of neurons = number of classes (classification) or 1 (regression)
- **Softmax output** is commonly used for multiclass classification:

\[
p_i = \frac{e^{w_i \cdot x'}}{\sum_{t=0}^{c-1} e^{w_t \cdot x'}}
\]

- Where:
  - \(c\) = number of output neurons (classes)  
  - \(x'\) = input vector from previous layer  
  - \(w_i\) = row of weight matrix corresponding to neuron i  

- Example: 3 output neurons, all receiving the same input \(x'\):

\[
p_0 = \frac{e^{w_0 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}
\]

---

### 3. Key Points

- Each **pixel** in an image is treated as a feature.  
- Input size grows with **image dimensions  channels**.  
- Softmax converts raw outputs of neurons into **probabilities**.  
- Weight matrices connect neurons between layers and determine transformations.


# Multiclass Classification: The Softmax Function

The **Softmax output** is a multiclass logistic function used to compute the 'probability' of an input belonging to one of multiple classes.

---

## 1. The General Softmax Equation
The function calculates the probability $p_i$ for a specific class $i$ out of $c$ total classes:

$$p_i = \frac{e^{w_i \cdot x'}}{\sum_{t=0}^{c-1} e^{w_t \cdot x'}}$$

---

## 2. Example: 3-Neuron Output Layer
If the output layer has **3 neurons** (representing classes 0, 1, and 2) receiving the same input $x'$, the probabilities are calculated as follows:

**For Class 0:**
$$p_0 = \frac{e^{w_0 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$$

**For Class 1:**
$$p_1 = \frac{e^{w_1 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$$

**For Class 2:**
$$p_2 = \frac{e^{w_2 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$$

---

## 3. Key Components of the Equation
* **$x'$ (Input Vector):** A large vector arriving from the previous layers of the network.
* **$w_i$ (Weight Matrix Rows):** Rows of the weight matrix located between layers.
* **$c$:** The total number of neurons (classes) in the output layer.
* **Numerator:** Represents the non-normalized probability (exponentiated score) for class $i$.
* **Denominator:** The sum of all exponentiated scores, ensuring the final probabilities sum to 1.

# Quiz 1: Identifying Softmax Expressions

### **Question**
Based on the Softmax function for a 3-neuron output layer (representing classes 0, 1, and 2), which of the following is the correct mathematical expression to calculate the probability $p_1$?

1. $\frac{e^{w_0 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$
2. $\frac{e^{w_1 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$
3. $\frac{e^{w_2 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$

---

### **Correct Answer**
**Option 2** is correct:
$$\Large p_1 = \frac{e^{w_1 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}}$$

---

### **Detailed Explanation**

The **Softmax function** follows a specific structure to convert raw model outputs (logits) into probabilities that sum up to 1:

1.  **The Numerator (Specific Class):** To find the probability of a specific class $i$, the numerator must use the exponentiated score of that specific neuron ($e^{w_i \cdot x'}$). Since we are looking for $p_1$, the numerator must be $e^{w_1 \cdot x'}$.
2.  **The Denominator (Normalization):** The denominator is the sum of the exponentiated scores for **all** neurons in the layer. This acts as a normalization constant. In a 3-neuron system, this sum is always $e^{w_0 \cdot x'} + e^{w_1 \cdot x'} + e^{w_2 \cdot x'}$, regardless of which specific probability ($p_0, p_1,$ or $p_2$) you are calculating.
3.  **Probability Interpretation:** By dividing the specific neuron's score by the total sum, we ensure that $0 \le p_i \le 1$ and that $p_0 + p_1 + p_2 = 1$.

# Quiz 2: Range of Softmax Probabilities

### **Question**
In a softmax output layer with three neurons ($p_0, p_1, p_2$), if the probability of the first class is fixed at **$p_0 = 0.5$**, what is the possible range of values for **$p_1$**?

1. 0 to 1
2. **0 to 0.5**
3. 0 to 0.25
4. 0.5 to 1

---

### **Correct Answer**
**Option 2: 0 to 0.5**

---

### **Detailed Explanation**

The Softmax function ensures that the sum of all output probabilities is exactly **1**. For a three-class system, the relationship is defined by the following equation:

$$p_0 + p_1 + p_2 = 1$$

If we know that **$p_0 = 0.5$**, we can substitute this into the equation:

$$0.5 + p_1 + p_2 = 1$$
$$p_1 + p_2 = 0.5$$

Because probabilities generated by the Softmax function can never be negative ($p_i \ge 0$), the value of $p_1$ is constrained by the remaining "probability budget":

* **Minimum value of $p_1$:** If $p_2$ takes the entire remaining value ($0.5$), then **$p_1 = 0$**.
* **Maximum value of $p_1$:** If $p_2$ is $0$, then **$p_1 = 0.5$**.

Therefore, $p_1$ must fall within the range of **0 to 0.5**.

# Inputs and Outputs of a Neural Network: Softmax vs. Sigmoid

### 1. Sigmoid as a Special Case of Softmax
Binary classification is a special case of multiclass classification. In a **sigmoid output**, only one neuron is needed in the output layer because for two classes ($p_0$ and $p_1$), the sum $p_0 + p_1 = 1$. Computing one value automatically gives you the other.

**Derivation:**
Starting with a 2-neuron softmax:
$$p_1 = \frac{e^{w_1 \cdot x'}}{e^{w_0 \cdot x'} + e^{w_1 \cdot x'}}$$

By dividing the numerator and denominator by $e^{w_1 \cdot x'}$, we get:
$$p_1 = \frac{1}{1 + e^{(w_0 - w_1) \cdot x'}}$$

Replacing $(w_1 - w_0)$ with a single weight $w$ results in the standard **Sigmoid function**.

## Quiz: Softmax Calculation

**Question:**
Given the following weight vectors and input $x'$:
* $w_0 = [1, 1, -1]$
* $w_1 = [2, 0, -1]$
* $w_2 = [1, 2, 2]$
* $x' = [2, 1, 1]^T$

What are the values of $p_0, p_1,$ and $p_2$ (up to 3 decimal points)?

### **Correct Answer**
**Option 2: 0.017, 0.047, 0.936**

### **Step-by-Step Explanation**

1. **Calculate Dot Products ($w_i \cdot x'$):**
   * $score_0 = (1 \times 2) + (1 \times 1) + (-1 \times 1) = 2 + 1 - 1 = \mathbf{2}$
   * $score_1 = (2 \times 2) + (0 \times 1) + (-1 \times 1) = 4 + 0 - 1 = \mathbf{3}$
   * $score_2 = (1 \times 2) + (2 \times 1) + (2 \times 1) = 2 + 2 + 2 = \mathbf{6}$

2. **Calculate Exponentials ($e^{score}$):**
   * $e^2 \approx 7.389$
   * $e^3 \approx 20.086$
   * $e^6 \approx 403.429$

3. **Calculate the Sum (Denominator):**
   * $Sum = 7.389 + 20.086 + 403.429 = \mathbf{430.904}$

4. **Calculate Final Probabilities ($p_i = \frac{e^{score_i}}{Sum}$):**
   * $p_0 = 7.389 / 430.904 \approx \mathbf{0.017}$
   * $p_1 = 20.086 / 430.904 \approx \mathbf{0.047}$
   * $p_2 = 403.429 / 430.904 \approx \mathbf{0.936}$

# Comparison: Sigmoid vs. Softmax Functions

In neural networks, choosing between Sigmoid and Softmax depends on whether you are performing binary or multiclass classification.

| Feature | Sigmoid Function | Softmax Function |
| :--- | :--- | :--- |
| **Primary Use** | Binary classification (2 classes). | Multiclass classification (> 2 classes). |
| **Output Layer** | Typically uses only **one neuron**. | Uses **multiple neurons** (one per class). |
| **Probability Sum** | $p_0 + p_1 = 1$; only one value needs calculation. | All probabilities $\sum_{i=0}^{n} p_i$ must equal **1**. |
| **Output Range** | Maps input to a range between **0 and 1**. | Converts scores into a **probability distribution**. |
| **Logic** | A special case of the Softmax function. | A generalization of the Sigmoid for multiple outputs. |

---

### Key Takeaways from the Derivation
* **Sigmoid from Softmax**: The sigmoid function can be derived by assuming a 2-neuron softmax and dividing the expression by the numerator.
* **Weight Relationship**: If we replace the weight difference $(w_1 - w_0)$ with a single vector $w$, we arrive at the sigmoid function.
* **MNIST Example**: In a dataset like MNIST (10 classes), the output layer has 10 neurons, each giving the probability of a specific digit. The image is classified to the neuron with the highest probability.

## Assumptions Made to Simplify Neural Networks
To manage the complexity of large neural networks and make them mathematically tractable and computationally efficient, several **standard architectural assumptions** are commonly adopted. These assumptions define how information flows through the network and how learning takes place.

---

### 1. Layered and Sequential Structure
- Neurons are organized into **layers**:  
  **Input Layer  Hidden Layer(s)  Output Layer**
- Information flows **only forward**, from one layer to the next.
- This structure is known as a **Feedforward Neural Network (FNN)**.
- It simplifies:
  - Network visualization
  - Gradient computation during backpropagation
  - Implementation in software frameworks

---

### 2. No Intra-Layer Communication
- Neurons **within the same layer do not communicate** with each other.
- Each neuron processes the same input vector independently.
- Benefits:
  - Enables **parallel computation**
  - Avoids cyclic dependencies within a layer
- Interaction only happens **across layers**, not within them.

---

### 3. Input and Output Layer Constraints
- **All inputs enter through the input layer only**
  - Each input neuron represents one feature/attribute
- **All outputs are produced by the output layer**
  - Classification  number of neurons = number of classes
  - Regression  usually a single neuron
- No hidden layer directly receives raw input or produces final output.

---

### 4. Dense Connectivity Between Consecutive Layers
- Each neuron in layer \( l \) is connected to **every neuron** in layer \( l+1 \).
- Known as a **Fully Connected (Dense) Layer**.
- Implications:
  - Maximum information flow
  - High representational power
  - Increased number of trainable parameters
- Trade-off:
  - Powerful but computationally expensive for large networks

---

### 5. Weights and Biases Everywhere
- **Every connection** has an associated **weight**
- **Every neuron** has an associated **bias**
- Mathematical form:
  \[
  z = w^T x + b
  \]
- Purpose:
  - Weights control feature importance
  - Bias allows activation even when input is zero
- Both weights and biases are **learned during training**

---

### 6. Same Activation Function per Layer
- All neurons in a given layer use the **same activation function**
- Common choices:
  - ReLU (hidden layers)
  - Sigmoid / Softmax (output layer)
- Advantages:
  - Uniform gradient behavior
  - Simpler backpropagation
  - Consistent feature transformation within a layer

---

### Why These Assumptions Matter
- Enable **efficient training** using backpropagation
- Reduce architectural complexity
- Make neural networks:
  - Scalable
  - Interpretable
  - Easier to implement and optimize

---

### Key Takeaway
These simplifying assumptions form the foundation of most **classical neural networks**. While advanced architectures (CNNs, RNNs, Transformers) relax some of these assumptions, understanding them is essential before exploring more complex models.


## Parameters and Hyperparameters of Neural Networks  In-Depth Summary

Training a neural network means **learning the optimal values of its parameters** so that the model can make accurate predictions on unseen data. Similar to traditional machine learning models, neural networks aim to minimize a **loss (cost) function**, but they do so with a more complex structure.

---

## 1. What Does Training a Neural Network Mean?
- Training is the process of **adjusting weights and biases** to minimize a loss function.
- The network learns patterns by comparing:
  - **Predicted output** vs **ground truth labels**
- Optimization is typically done using:
  - Gradient Descent
  - Backpropagation

---

## 2. Parameters of a Neural Network (Learned During Training)

###  Weights (\( W \))
- Represent the **strength of connections** between neurons.
- Stored as **weight matrices** between layers.
- Updated iteratively during training.

###  Biases (\( b \))
- Added to weighted sums before activation.
- Allow neurons to activate even when inputs are zero.
- One bias per neuron.

 **Parameters = Weights + Biases**  
These are learned automatically from data.

---

## 3. Hyperparameters of a Neural Network (Set Before Training)

Hyperparameters are **not learned**; they are defined by the network designer.

### Common Hyperparameters:
- Number of layers (depth)
- Number of neurons in each layer
- Activation functions (ReLU, Sigmoid, Softmax, etc.)
- Learning rate
- Batch size
- Number of epochs
- Network architecture (fully connected, CNN, etc.)

 Hyperparameters control **how learning happens**, not what is learned.

---

## 4. Difference Between Parameters and Hyperparameters

| Aspect | Parameters | Hyperparameters |
|------|------------|----------------|
| Learned from data |  Yes |  No |
| Set before training |  No |  Yes |
| Examples | Weights, Biases | Layers, Neurons, Activations |
| Updated during training |  Yes |  No |

---

## 5. Learning Process Overview
1. Fix the **hyperparameters**
2. Initialize weights and biases
3. Forward pass  compute predictions
4. Compute loss
5. Backpropagate error
6. Update weights and biases
7. Repeat until convergence

---

# <font size="6">Activation Functions in Neural Networks</font>

### <font size="5">What is an Activation Function?</font>
An **activation function** determines the output of a neuron by applying a non-linear transformation to the weighted sum of inputs plus bias.

If the cumulative input to a neuron is:
$$y = w_1 a_1 + w_2 a_2 + \dots + w_k a_k + b$$

Then the neurons output is:
$$a = f(y)$$
where $f(\cdot)$ is the activation function.

---

### <font size="5">Key Properties of Activation Functions</font>
Activation functions should ideally satisfy the following:

* **Smoothness**: They should not have abrupt jumps, making them differentiable and suitable for gradient-based optimization.
* **Non-linearity**: Non-linear functions allow neural networks to learn complex patterns. Without non-linearity, deep networks collapse into simple linear models.

---

### <font size="5">Common Activation Functions</font>

#### 1. Logistic (Sigmoid) Function
$$f(x) = \frac{1}{1 + e^{-x}}$$



**Properties:**
* **Output range**: $(0, 1)$
* Often used in binary classification
* Smooth and differentiable
* Can suffer from **vanishing gradients** for large $|x|$
* **Neuron output with sigmoid**: $$a = \frac{1}{1 + e^{-(w_1 a_1 + \dots + w_k a_k + b)}}$$

#### 2. Hyperbolic Tangent (tanh)
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$



**Properties:**
* **Output range**: $(-1, 1)$
* **Zero-centered**: generally better than sigmoid for optimization
* Still suffers from vanishing gradients

#### 3. Rectified Linear Unit (ReLU)
$$f(x) = \begin{cases} x, & x \geq 0 \\ 0, & x < 0 \end{cases}$$



**Properties:**
* Computationally efficient
* Helps mitigate the vanishing gradient problem
* Can suffer from **dying ReLU** (neurons stuck at zero)

#### 4. Leaky ReLU
$$f(x) = \begin{cases} x, & x \geq 0 \\ \alpha x, & x < 0 \end{cases}$$
where $\alpha$ is a small constant (e.g., $0.01$).

**Properties:**
* Fixes the dying ReLU problem
* Allows small negative outputs

---

### <font size="5">Relationship Between Sigmoid Neuron and Perceptron</font>
If all weights and biases of a sigmoid neuron are multiplied by a positive constant $c > 0$:
$$a = \frac{1}{1 + e^{-c(w \cdot x + b)}}$$

Then, as $c \to \infty$:
* The sigmoid behaves like a **step function**.
* The neuron behaves exactly like a **perceptron**.
* This holds when $w \cdot x + b \neq 0$ for the input $x$.

---

### <font size="5">Key Takeaway</font>
* Activation functions introduce **non-linearity**, enabling neural networks to model complex relationships.
* Sigmoid, tanh, ReLU, and Leaky ReLU are among the most widely used.
* Without activation functions, deep neural networks would lose their expressive power.

$$y = (w_1x_1 + w_2x_2 + w_3x_3) + b$$
$$y = (2 \times 3) + (-6 \times 2) + (3 \times 1) + (-1)$$
$$y = 6 - 12 + 3 - 1$$
$$y = -4$$

**Step 2: Apply the Leaky ReLU Activation Function**
The Leaky ReLU function is defined as:
* $f(x) = x$ for $x \geq 0$
* $f(x) = \alpha x$ for $x < 0$

Since our cumulative input $y = -4$ is negative ($x < 0$), we use the second part of the function with the given $\alpha = 0.2$:
$$f(-4) = 0.2 \times (-4)$$
$$f(-4) = -0.8$$

**Final Result:**
The output of the neuron is **-0.800**.

Unlike the standard ReLU which would have blocked this signal entirely (outputting 0), Leaky ReLU allows a small negative value to pass through, which helps prevent neurons from becoming "dead" during training.

# Summary of Neural Networks Session

In this session, you learned:

1. **Inspiration from the Human Brain:**
   - Artificial Neural Networks (ANNs) are modeled after the structure and functioning of biological neurons.

2. **Perceptron as a Binary Classifier:**
   - A perceptron can classify data points into two classes.
   - Multiple perceptrons can perform more complex classification tasks, including multiclass classification.

3. **ANN Architecture:**
   - **Topology:** Neurons arranged in layers (input, hidden, output).
   - **Parameters:** Weights (W) and biases (b) that the network learns during training.
   - **Hyperparameters:** Predefined values like number of layers, neurons per layer, activation functions, etc.

4. **Simplifying Assumptions:**
   - Neurons are arranged sequentially in layers.
   - Neurons in the same layer do not interact.
   - All inputs enter through the input layer; outputs exit through the output layer.
   - Layers are densely connected (all-to-all).
   - Every neuron has an associated bias and uses the same activation function within a layer.

5. **Numerical Inputs:**
   - ANNs require numeric inputs.
   - Non-numeric data (text, categorical features) must be converted to numeric form (e.g., one-hot encoding, word embeddings).

6. **Notations Fixed:**
   - `W`: weight matrix
   - `b`: bias
   - `x`: input vector
   - `y`: ground truth label
   - `p`: probability vector of predicted output
   - `h`: output of hidden layers
   - Superscripts: denote layer number
   - Subscripts: denote neuron index

7. **Activation Functions:**
   - Common functions used to introduce non-linearity:
     - **Sigmoid:** smooth S-shaped curve, outputs 0 to 1.
     - **Tanh:** outputs -1 to 1, zero-centered.
     - **ReLU:** outputs 0 for negative inputs, linear for positive.
     - **Leaky ReLU:** allows small negative slope to avoid dead neurons.

> These concepts lay the foundation for understanding how ANNs transform inputs through layers to produce outputs and learn patterns from data.


# Key Takeaways from Neural Networks Session

1. **Perceptrons Form the Building Blocks:**  
   - A perceptron is a simple binary classifier, and multiple perceptrons can be combined to solve complex or multiclass classification tasks.

2. **ANN Architecture and Learning:**  
   - ANNs consist of layers of neurons with weights and biases as parameters. Hyperparameters like number of layers, neurons, and activation functions define the network's structure.

3. **Non-linearity and Numerical Inputs:**  
   - Activation functions (sigmoid, ReLU, tanh, etc.) introduce non-linearity, enabling the network to learn complex patterns. All inputs must be numeric for the network to process them.
