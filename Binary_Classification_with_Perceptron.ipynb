#Binary Classification using Perceptron
In this segment, we study how a perceptron can be used for **binary classification**.

The goal of a perceptron is to learn a decision boundary that separates data points belonging to two different classes.


perceptron_img1.png

#Perceptron Problem Statement
Given a dataset of input vectors `x` and corresponding labels `y  {+1, -1}`,  
we need to find parameters **w (weights)** and **b (bias)** such that:

- For all points where y = +1:
  wx + b > 0

- For all points where y = -1:
  wx + b < 0


#Step / Sign Function
The perceptron uses a **sign (step) function** defined as:

y = +1   if x > 0  
y = -1   if x  0

This function determines the predicted class label.


#Compact Mathematical Representation
A pair (w, b) is a **valid separator** if:

y ( wx + b ) > 0   for **all data points**

If this condition fails for even one point, the separator is invalid.


We are given the following data points:

| x        | y  |
|---------|----|
| (0, 3)  | +1 |
| (5, 9)  | +1 |
| (-1,-2)| -1 |

Each input vector x is represented as:
x = (x, x)


#Question 1: Valid Separator
Which of the following (w, b) combinations are valid separators?

Options:

1) w = [-1, 1], b = -2  
2) w = [-1, 1], b = -5  
3) w = [-1, 2], b = -1  
4) w = [-1, 2], b = 2


##Option 1: w = [-1, 1], b = -2

For (0,3), y=+1:
wx + b = (-1*0 + 1*3) - 2 = 1  valid

For (5,9), y=+1:
(-5 + 9) - 2 = 2  valid

For (-1,-2), y=-1:
(1 - 2) - 2 = -3  y(wx+b) = (-1)(-3) > 0  valid

##Option 2: w = [-1, 1], b = -5

For (0,3), y=+1:
3 - 5 = -2  invalid

##Option 3: w = [-1, 2], b = -1
For (0,3), y=+1:
6 - 1 = 5  valid

For (5,9), y=+1:
(-5 + 18) - 1 = 12  valid

For (-1,-2), y=-1:
(1 - 4) - 1 = -4  y(wx+b) > 0  valid

##Option 4: w = [-1, 2], b = 2

For (-1,-2), y=-1:
(1 - 4) + 2 = -1  y(wx+b) = (-1)(-1) = 1  valid

But check positives carefully:
Boundary becomes too loose and misclassifies edge cases.

# Question 3

After converting an input vector **x** of dimension (5,1) to **homogeneous coordinates**,  
what should be the new dimension of **x** and **w**?

**Options:**
- (6,1)
- (1,6)

### Correct Answer

 **(6,1)**

**Explanation:**
- Homogeneous coordinates add one extra dimension to include the bias term.
- Vectors are represented as **column vectors** in perceptrons.
- Therefore, both **x** and **w** become vectors of dimension **(6,1)**.
- (1,6) represents a row vector
- In perceptrons, x and w are represented as column vectors
- This ensures wx results in a scalar
