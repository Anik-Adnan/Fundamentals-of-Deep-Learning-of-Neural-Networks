# Introduction to Perceptron

perceptron.jpeg

A **perceptron** is a simple computational model that mimics human decision-making.  
It takes multiple inputs, multiplies them by their corresponding weights, adds a bias, and applies a **step function** to decide whether to "fire" (1/Yes) or "stay dormant" (0/No).  

This was the first step towards building neural networks.


#Real-life Analogy

Imagine deciding whether to visit a sushi place on Saturday. Factors influencing your decision:

- Distance of the place
- Cost of visiting
- Number of friends accompanying you

Each factor has a different **importance (weight)** in your decision.  
The perceptron works in a similar way.


#Step Function

The **step function** defines the perceptron’s output:

y = 1 if x > 0  
y = 0 if x ≤ 0  

Here, x is the cumulative input:

Cumulative Input = w1*x1 + w2*x2 + ... + wk*xk + b

Where:
- w = vector of weights
- x = vector of inputs
- b = bias


#Vector Representation

Using vectors, the weighted sum is:

w^T . x + b

Where:

w = [w1, w2, ..., wk]^T (column vector)  
x = [x1, x2, ..., xk]^T (column vector)  
b = bias  

This simplifies computation and allows scaling to larger networks.


#Perceptron Output

We have a perceptron with the following parameters:

- Weight vector:

w =
[3, 1, 5, 7, 4, 0]^T  

- Input vector:

x =
[1, 1, 0, 1, 0, 0]^T  

- Bias: b = -2  

The perceptron output is calculated as:

Cumulative Input = w^T . x + b

Then apply the **step function**:

y = 1 if Cumulative Input > 0  
y = 0 if Cumulative Input ≤ 0


```python
import numpy as np

# Define vectors
w = np.array([3, 1, 5, 7, 4, 0])
x = np.array([1, 1, 0, 1, 0, 0])
b = -2

# Compute cumulative input
cumulative_input = np.dot(w, x) + b
print("Cumulative Input:", cumulative_input)

# Apply step function
output = 1 if cumulative_input > 0 else 0
print("Perceptron Output:", output)
```

#Another example
We have a perceptron with:

- Weight vector: w = [3, 1, 5, 7, 4, 0]  
- Input vector: x = [1, 1, 0, 1, 0, 0]  
- Bias: b = -2

The **cumulative input** is calculated as:

Cumulative Input = w^T · x + b

1. Compute the dot product w^T · x:

w^T · x = (3*1) + (1*1) + (5*0) + (7*1) + (4*0) + (0*0)
         = 3 + 1 + 0 + 7 + 0 + 0
         = 11

2. Add the bias:

Cumulative Input = 11 + (-2) = 9


```python
import numpy as np

w = np.array([3, 1, 5, 7, 4, 0])
x = np.array([1, 1, 0, 1, 0, 0])
b = -2

cumulative_input = np.dot(w, x) + b
print("Cumulative Input:", cumulative_input)
```

#Perceptron Decision Example

We have the following:

### Input vector:
x = [1, 0, 1]  # Distance = 1, Cost = 0, Company = 1

### Sample weights (sum = 1):
w = [0.5, 0.3, 0.2]

### Bias:
b = -0.7

The perceptron decision is computed as:

Cumulative Input = w^T · x + b

Then apply the step function:

y = 1 if Cumulative Input > 0 → "Yes"  
y = 0 if Cumulative Input ≤ 0 → "No"


```python
import numpy as np

# Input vector
x = np.array([1, 0, 1])

# Weights and bias
w = np.array([0.5, 0.3, 0.2])
b = -0.7

# Compute cumulative input
cumulative_input = np.dot(w, x) + b
print("Cumulative Input:", cumulative_input)

# Apply step function
decision = "Yes" if cumulative_input > 0 else "No"
print("Perceptron Decision:", decision)
```
